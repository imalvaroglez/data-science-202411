## MÃ©tricas

Los modelos deben evaluarse para asegurar su precisiÃ³n y confiabilidad. Una herramienta muy utilizada es la matriz de confusiÃ³n.

La matriz de confusiÃ³n resume la eficiencia del modelo sobre el conjunto de prueba, indicando las instancias predecidas correctamente y las que no:

* True Positive (TP): El modelo predijo correctamente una salida positiva
* True Negative (TN): El modelo predijo correctamente una salida negativa
* False Positive (FP): El modelo predijo incorrectamente una salida positiva
* False Negative (FN): El modelo predijo incorrectamente una salida negativa
